```python
import os

def print_files_in_dir(root_dir, prefix):
    files = os.listdir(root_dir)
    for file in files:
        path = os.path.join(root_dir, file)
        print(prefix + path)
        if os.path.isdir(path):
            print_files_in_dir(path, prefix + "    ")
```

```python
root_dir = '../../../../../Documents/pshopper/mmfashion/data/Attr_Predict/Img_fortest'
print_files_in_dir(root_dir, "")
```

```python
root_dir = '../../../../../Documents/pshopper/mmfashion/data/Attr_Predict/Img'


import os

dict = {}

dirs = os.listdir(root_dir)
for dir in dirs:
    path = os.path.join(root_dir, dir)
    
    if os.path.isdir(path):
        file_list = os.listdir(path)
        name = path.split('\\')[-1]
        dict[name] = len(file_list)

print(dict)
```
